```python
import pandas as pd
red_df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv',sep = ';',header = 0,engine = 'python')
white_df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-white.csv',sep = ';',header = 0,engine = 'python')
```

```python
red_df.insert(0, column = 'type', value = 'red')
red_df.head()
```

```python
red_df.shape
```

```python
white_df.head()
```

```python
white_df.insert(0,'type','white')
white_df.head()
```

```python
white_df.shape
```

```python
wine = pd.concat([red_df,white_df])
wine.shape
```

```python
print(wine.info())
```

```python
from google.colab import files
wine.to_csv('wine.csv')
files.download('wine.csv')
```

```python
wine.columns = wine.columns.str.replace(' ','_')#컬럼명의 공백을 _ 로 변경 추후 오류 방지
wine.head()
```

```python
wine.describe()
```

```python
sorted(wine.quality.unique())
```

```python
wine.quality.value_counts()
```

```python
wine.groupby('type')['quality'].describe()
```

```python
wine.groupby('type')['quality'].mean()
```

```python
wine.groupby('type')['quality'].std()
```

```python
wine.groupby('type')['quality'].agg(['mean','std']) # mean()과 std()를 묶어서 사용하기 위해
```
